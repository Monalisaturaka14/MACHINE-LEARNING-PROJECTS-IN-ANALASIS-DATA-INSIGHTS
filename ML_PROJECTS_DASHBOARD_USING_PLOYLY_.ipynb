```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
from ipywidgets import Layout
import warnings

warnings.filterwarnings('ignore')

```

```python
projects_data = [
    {
        'id': 1,
        'name': 'Email Spam Detection',
        'category': 'NLP',
        'type': 'Classification',
        'accuracy': 97,
        'precision': 95,
        'recall': 94,
        'f1_score': 94.5,
        'description': 'Spam classification using NLP and ML algorithms.',
        'algorithms': ['TF-IDF', 'Naive Bayes', 'Random Forest', 'Logistic Regression'],
        'dataset_size': 5000,
        'features': 3000,
        'technologies': ['Python', 'Scikit-learn', 'NLTK', 'Pandas']
    },
    {
        'id': 2,
        'name': 'Heart Disease Prediction',
        'category': 'Healthcare',
        'type': 'Classification',
        'accuracy': 85,
        'precision': 83,
        'recall': 87,
        'f1_score': 85,
        'description': 'Heart disease risk prediction using ML.',
        'algorithms': ['Logistic Regression', 'Random Forest', 'Gradient Boosting', 'SVM'],
        'dataset_size': 1000,
        'features': 13,
        'technologies': ['Python', 'Scikit-learn', 'Pandas', 'Matplotlib']
    },
    {
        'id': 3,
        'name': 'House Price Prediction',
        'category': 'Real Estate',
        'type': 'Regression',
        'r2_score': 85,
        'rmse': 25000,
        'mae': 18000,
        'description': 'Property price prediction using regression models.',
        'algorithms': ['Linear Regression', 'Ridge', 'Lasso', 'Random Forest'],
        'dataset_size': 1000,
        'features': 13,
        'technologies': ['Python', 'Scikit-learn', 'Pandas', 'Seaborn']
    }
]

df_projects = pd.DataFrame(projects_data)

print("Projects loaded:", len(df_projects))
df_projects
```

```python
def display_stats():
    total_projects = len(df_projects)
    classification_df = df_projects[df_projects['type'] == 'Classification']
    avg_accuracy = classification_df['accuracy'].mean()
    total_samples = df_projects['dataset_size'].sum()

    print("Total Projects:", total_projects)
    print("Average Accuracy:", round(avg_accuracy, 2))
    print("Total Data Samples:", total_samples)

display_stats()
```

```python
def create_accuracy_chart():
    classification_df = df_projects[df_projects['type'] == 'Classification']

    fig = go.Figure()

    fig.add_trace(go.Bar(
        x=classification_df['name'],
        y=classification_df['accuracy'],
        text=classification_df['accuracy'],
        textposition='auto'
    ))

    fig.update_layout(
        title='Classification Projects - Accuracy Comparison',
        xaxis_title='Project',
        yaxis_title='Accuracy (%)',
        template='plotly_white'
    )

    fig.show()

create_accuracy_chart()
```

```python
def create_category_distribution():
    category_counts = df_projects['category'].value_counts()

    fig = go.Figure(data=[go.Pie(
        labels=category_counts.index,
        values=category_counts.values,
        hole=0.4
    )])

    fig.update_layout(
        title='Projects by Category',
        template='plotly_white'
    )

    fig.show()

create_category_distribution()
```

```python
def create_dataset_chart():
    fig = go.Figure(data=[go.Bar(
        x=df_projects['name'],
        y=df_projects['dataset_size'],
        text=df_projects['dataset_size'],
        textposition='auto'
    )])

    fig.update_layout(
        title='Dataset Sizes Across Projects',
        xaxis_title='Project',
        yaxis_title='Samples',
        template='plotly_white'
    )

    fig.show()

create_dataset_chart()
```

```python
def create_technology_heatmap():
    all_techs = list(set([tech for proj in projects_data for tech in proj['technologies']]))
    tech_matrix = []

    for proj in projects_data:
        row = [1 if tech in proj['technologies'] else 0 for tech in all_techs]
        tech_matrix.append(row)

    fig = go.Figure(data=go.Heatmap(
        z=tech_matrix,
        x=all_techs,
        y=[proj['name'] for proj in projects_data]
    ))

    fig.update_layout(
        title='Technology Stack Across Projects',
        template='plotly_white'
    )

    fig.show()

create_technology_heatmap()
```

```python
def display_project_details(project_id):
    project = df_projects[df_projects['id'] == project_id].iloc[0]
    print("Project:", project['name'])
    print("Category:", project['category'])
    print("Type:", project['type'])
    print("Dataset Size:", project['dataset_size'])
    print("Features:", project['features'])
    print("Algorithms:", project['algorithms'])
    print("Technologies:", project['technologies'])
    print("Description:", project['description'])

display_project_details(1)
```

```python
def create_precision_recall_scatter():
    classification_df = df_projects[df_projects['type'] == 'Classification']

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=classification_df['precision'],
        y=classification_df['recall'],
        mode='markers+text',
        text=classification_df['name'],
        textposition='top center',
        marker=dict(
            size=classification_df['accuracy'],
            color=classification_df['accuracy'],
            colorscale='Viridis',
            showscale=True
        )
    ))

    fig.update_layout(
        title='Precision vs Recall Comparison',
        xaxis_title='Precision (%)',
        yaxis_title='Recall (%)',
        template='plotly_white'
    )

    fig.show()

create_precision_recall_scatter()
```

```python
def create_radar_chart():
    classification_df = df_projects[df_projects['type'] == 'Classification']

    fig = go.Figure()

    for _, row in classification_df.iterrows():
        fig.add_trace(go.Scatterpolar(
            r=[row['accuracy'], row['precision'], row['recall']],
            theta=['Accuracy', 'Precision', 'Recall'],
            fill='toself',
            name=row['name']
        ))

    fig.update_layout(
        polar=dict(radialaxis=dict(visible=True, range=[70, 100])),
        title="Classification Performance Radar",
        showlegend=True
    )

    fig.show()

create_radar_chart()
```

```python
from collections import Counter

def create_algorithm_frequency():
    all_algorithms = []
    for proj in projects_data:
        all_algorithms.extend(proj['algorithms'])

    algo_counts = Counter(all_algorithms)

    fig = go.Figure(go.Bar(
        x=list(algo_counts.values()),
        y=list(algo_counts.keys()),
        orientation='h'
    ))

    fig.update_layout(
        title='Algorithm Usage Frequency',
        xaxis_title='Number of Projects',
        yaxis_title='Algorithm',
        template='plotly_white'
    )

    fig.show()

create_algorithm_frequency()
```

```python
def create_size_vs_features():
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=df_projects['dataset_size'],
        y=df_projects['features'],
        mode='markers+text',
        text=df_projects['name'],
        textposition='top center',
        marker=dict(size=12)
    ))

    fig.update_layout(
        title='Dataset Size vs Number of Features',
        xaxis_title='Dataset Size',
        yaxis_title='Number of Features',
        template='plotly_white'
    )

    fig.show()

create_size_vs_features()
```

```python
def create_metrics_correlation():
    classification_df = df_projects[df_projects['type'] == 'Classification']

    metrics = classification_df[['accuracy', 'precision', 'recall']]
    corr = metrics.corr()

    fig = go.Figure(data=go.Heatmap(
        z=corr.values,
        x=corr.columns,
        y=corr.columns
    ))

    fig.update_layout(
        title='Correlation Between Metrics',
        template='plotly_white'
    )

    fig.show()

create_metrics_correlation()
```

```python
def create_project_type_count():
    type_counts = df_projects['type'].value_counts()

    fig = go.Figure(go.Bar(
        x=type_counts.index,
        y=type_counts.values
    ))

    fig.update_layout(
        title='Project Types Distribution',
        xaxis_title='Project Type',
        yaxis_title='Count',
        template='plotly_white'
    )

    fig.show()

create_project_type_count()
```
